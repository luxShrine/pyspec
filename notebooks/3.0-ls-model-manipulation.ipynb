```python
# store data
from dataclasses import asdict
import json

class json_serialize(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return json.JSONEncoder.default(self, obj)

list_oof_stats_d = [s.__dict__ for s in list_oof_stats]
list_oof_stats_j = json.dumps(list_oof_stats_d, cls=json_serialize) 
ce_j = json.dumps(asdict(ce))

Path("oofstats.json").write_text(list_oof_stats_j)
Path("ce.json").write_text(ce_j)

```

```python

from torch import nn
class LSM2(nn.Module):
    def __init__(self, c: int):
        super().__init__()
        self.proj = nn.Conv2d(c, c, kernel_size=1, bias=True)

        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.proj(x)

    def get_loss_managers(
        self, lr: float = 5e-4, wd: float = 1e-4, type: str = "LBFGS"
    ) -> tuple[nn.MSELoss, torch.optim.Optimizer]:
        """Create the optimizer and loss function directly from object."""
        if type == "LBFGS":
            return (
                nn.MSELoss(),
                torch.optim.LBFGS(self.parameters(), lr=1e-2, max_iter=100, history_size=10),
            )
        elif type == "AdamW":
            return (nn.MSELoss(), torch.optim.AdamW(self.parameters(), lr=lr, weight_decay=wd))
        else:
            raise RuntimeError
class LSM3(nn.Module):
    def __init__(self, c: int):
        super().__init__()
        self.proj = nn.Sequential(
            nn.Conv2d(c, c, kernel_size=1, bias=True),
            torch.nn.ReLU(),
            #torch.nn.Flatten(),
            #torch.nn.Linear(950, 1)
        )

        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.proj(x)

    def get_loss_managers(
        self, lr: float = 5e-4, wd: float = 1e-4, type: str = "LBFGS"
    ) -> tuple[nn.MSELoss, torch.optim.Optimizer]:
        """Create the optimizer and loss function directly from object."""
        if type == "LBFGS":
            return (
                nn.MSELoss(),
                torch.optim.LBFGS(self.parameters(), lr=1e-2, max_iter=100, history_size=10),
            )
        elif type == "AdamW":
            return (nn.MSELoss(), torch.optim.AdamW(self.parameters(), lr=lr, weight_decay=wd))
        else:
            raise RuntimeError
```
